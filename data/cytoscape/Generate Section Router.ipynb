```python
import json

def update_route(line, routing_graph):
    with open('../TrackModel-section-routing.json', 'r') as input:
        current_routing = json.load(input)

    current_routing[line] = routing_graph

    with open('../TrackModel-section-routing.json', 'w') as output:
        json.dump(current_routing, output)
        
def generate_routing_graph(graph_file):
    lines = graph_file.readlines()
    lines = list(map(lambda x: x.strip(), lines))
    edges = list(map(lambda x: x.split(' ', maxsplit=2), lines))
    
    nodes = set()
    edges_list = []
    
    edge_names = []
    
    for edge in edges:
        source, dest, name = edge
        
        nodes.add(source)
        nodes.add(dest)
        
        # Strip quotes
        name = name.strip('"')

        # Detect edge type
        edge_type = 'station'        
        if name.startswith('(') and name.endswith(')'):
            edge_type = 'in-between'
            
        edge_names.append(name)
            
        # generate cyto edge
        edge = {
            'data': {
                'id': name,
                'source': source,
                'target': dest
            }
        }
        
        if edge_type == 'station':
            edge['data']['station'] = True
            edge['data']['station_name'] = name
        
        edges_list.append(edge)
        
    nodes_list = []
    for node in nodes:
        nodes_list.append({
            'data': {
                'id': node,
            }
        })
    
    cyto = {
        'edges': edges_list,
        'nodes': nodes_list
    }
    
    print('\n'.join(edge_names))
    
    return cyto
```

```python
with open('green-line-routing-graph.txt', 'r') as green_line:
    green_routing_graph = generate_routing_graph(green_line)
    
update_route('green', green_routing_graph)
```

```python
with open('red-line-routing/red-line-routing.graph', 'r') as red_line:
    red_routing_graph = generate_routing_graph(red_line)
    
update_route('red', red_routing_graph)
```
