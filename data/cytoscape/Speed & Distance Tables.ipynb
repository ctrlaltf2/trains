# Green Line

```python
import pandas as pd

green_line_info = pd.read_csv('./TrackLayoutGreen.csv')
green_line_info

g_track_info = {}
```

```python
import json
from sympy import Rational, N

green_track_info = {}

for index, block in green_line_info.iterrows():
    block_id = block['Block ID']
    green_track_info[block_id] = {
        'length': float(block['Length']), # m
        'grade': float(block['Grade']),  # %
        'speed_limit': float(block['Speed Limit']), # km/hr
        'elevation': float(block['Elevation']), # ?
        'traverse_time': float(N( float(block['Length']) / (float(block['Speed Limit']) * Rational(5, 18)))), # s
    }
    
g_track_info['green'] = green_track_info
```

# Red Line

```python
import pandas as pd

red_line_info = pd.read_csv('./TrackLayoutRed.csv')
red_line_info
```

```python
import json
from sympy import Rational, N

red_track_info = {}

for index, block in red_line_info.iterrows():
    block_id = block['Block ID']
    red_track_info[block_id] = {
        'length': float(block['Length']), # m
        'grade': float(block['Grade']),  # %
        'speed_limit': float(block['Speed Limit']), # km/hr
        'elevation': float(block['Elevation']), # ?
        'traverse_time': float(N( float(block['Length']) / (float(block['Speed Limit']) * Rational(5, 18)))), # s
    }

def update_table(line, new_table):
    with open('../TrackModel-info.json', 'r') as input:
        current_table = json.load(input)

    current_table[line] = new_table

    with open('../TrackModel-info.json', 'w') as output:
        json.dump(current_table, output)
        
update_table('green', green_track_info)
update_table('red', red_track_info)
```
